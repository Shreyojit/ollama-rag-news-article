```python
# Import required libraries
import polars as pl
import ollama
import time


# Function to generate embeddings for the 'Article' column
def generate_embeddings_for_articles(df, text_column):
    """
    Generates embeddings for each row in the specified text column of a Polars DataFrame.

    Args:
    df (pl.DataFrame): The Polars DataFrame containing the text data.
    text_column (str): The name of the column containing the text.

    Returns:
    list: A list of embedding vectors corresponding to the rows in the DataFrame.
    """
    embeddings = []
    for i, text in enumerate(df[text_column].to_list(), start=1):
        start_time = time.time()
        try:
            # Call Ollama to generate embeddings for the text
            response = ollama.embeddings(model='nomic-embed-text', prompt=text)
            embeddings.append(response)
            print(f"Processed article {i}/{len(df)}: Success")
        except Exception as e:
            print(f"Error processing article {i}: {e}")
            embeddings.append(None)

        end_time = time.time()
        print(f"Time taken for article {i}: {end_time - start_time:.2f} seconds")

    return embeddings

```

```python
# Import required libraries
import polars as pl
import ollama
import time


# Function to generate embeddings for the 'Article' column
def generate_embeddings_for_articles(df, text_column):
    """
    Generates embeddings for each row in the specified text column of a Polars DataFrame.

    Args:
    df (pl.DataFrame): The Polars DataFrame containing the text data.
    text_column (str): The name of the column containing the text.

    Returns:
    list: A list of embedding vectors corresponding to the rows in the DataFrame.
    """
    embeddings = []
    for i, text in enumerate(df[text_column].to_list(), start=1):
        start_time = time.time()
        try:
            # Call Ollama to generate embeddings for the text
            response = ollama.embeddings(model='nomic-embed-text', prompt=text)
            
            # Assuming the response is in the form of a dictionary with an 'embedding' key
            embedding = response.get('embedding', [])
            embeddings.append(embedding)
            
            print(f"Processed article {i}/{len(df)}: Success")
        except Exception as e:
            print(f"Error processing article {i}: {e}")
            embeddings.append(None)

        end_time = time.time()
        print(f"Time taken for article {i}: {end_time - start_time:.2f} seconds")

    return embeddings

# Assuming articles is your Polars DataFrame and the 'Article' column contains text data
vectors = generate_embeddings_for_articles(articles, 'Article')

# Print the first embedding vector to check the format
print(vectors[0])  # This will print the first vector in the list
```

```python
import ollama

def generate_embedding_for_query(query, model='nomic-embed-text'):
  """
  Generates an embedding for a given query string using the specified Ollama model.

  Args:
    query: The input query string.
    model: The name of the Ollama model to use for embedding generation. 
            Defaults to 'nomic-embed-text'.

  Returns:
    A list representing the embedding vector for the query.
    Returns None if embedding generation fails.
  """
  try:
    response = ollama.embeddings(model=model, prompt=query)
    return response.get('embedding', None)
  except Exception as e:
    print(f"Error generating embedding for query '{query}': {e}")
    return None

# Example usage:
query = "public transport fares by 7 per cent"
query_embedding = generate_embedding_for_query(query)

if query_embedding:
  print("Embedding generated successfully:")
  print(query_embedding)
```

```python
collection.query(query_embeddings=query_embedding,n_results=3)
```
